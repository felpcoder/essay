# Informações sobre o dataframe

- A coluna competence trás como valor uma lista com cada uma das notas para cada uma das cinco competências, a soma das notas por competência é igual a coluna score. As cinco competências são: 

    * Adesão à norma culta da língua portuguesa.
    * Conformidade com o gênero textual argumentativo e com o tema proposto (enunciado), desenvolvendo um texto, utilizando conhecimentos de diferentes áreas.
    * Seleção, relação, organização e interpretação de dados e argumentos em defesa de um ponto de vista.
    * Uso de estruturas linguísticas argumentativas.
    * Elaboração de uma proposta para resolver o problema em questão.

    Para cada uma das competências as notas variam de 0 a 200, aumentando de 40 em 40 conforme tabela:
